```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
df.drop(columns=["class" , "deck" , "embark_town" , "alive" , "alone" , "class" , "who" , "adult_male"] , inplace=True)
```

```python
df
```

```python
X_Train , X_Test , y_Train , y_Test = train_test_split(df.drop(columns=["survived"]) , df["survived"] ,test_size=0.2 , random_state= 42)
```

```python
X_Train
```

```python
print(f"X_Test: {X_Test.shape[0]} samples , {X_Test.shape[1]} features")
```

Imputation transformer

```python
from sklearn.impute import SimpleImputer
```

```python
trf1 = ColumnTransformer([
    ("impute_age" , SimpleImputer() , [2]),
    ("impute_embarked", SimpleImputer(strategy = "most_frequent") , [6])] , remainder="passthrough")
```

```python
trf1
```

One hot Encoding 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
trf2 = ColumnTransformer([
    ("ohe_sex_embarked" , OneHotEncoder(sparse_output=False , handle_unknown="ignore") , [1,6])] ,  remainder = "passthrough")
```

```python
trf2
```

Scaling


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
trf3 = ColumnTransformer([
    ("scale" , MinMaxScaler(), slice(0,10))
])
```

```python
trf3
```

# feature Selection

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
trf4 = SelectKBest(score_func = chi2 , k = 8)
```

```python
print(trf4)
```

train the model 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
trf5 = DecisionTreeClassifier()
```

```python
# trf5.fit(X_Train, y_Train)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipeline = make_pipeline(trf1 , trf2 , trf3 , trf4 , trf5)
```

```python
# pipeline
```

```python
pipeline.fit(X_Train , y_Train)
```

```python
from sklearn.metrics import accuracy_score
```

Cross Validation using Pipeline

```python
from sklearn.model_selection import cross_val_score

cross_val_score(pipeline , X_Train , y_Train , cv=5 , scoring='accuracy').mean()
```

GridSearch Using pipeline

```python
# pipeline.named_steps
pipeline.get_params().keys()
```

```python
# params = {
#     'trf5__max_depth' : [1,2,3,4,5,None]
# }


params = {
    'decisiontreeclassifier__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipeline , params , cv=5 , scoring="accuracy")
grid.fit(X_Train , y_Train)
```
