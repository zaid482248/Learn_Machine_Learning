```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder 
```

```python
df = pd.read_csv(r"E:\Learn_ML\Data_sets\covid_toy.csv")
```

```python
print(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , y_Train , y_Test = train_test_split(df.drop(columns=["has_covid"]), df["has_covid"], test_size=0.2) 
```

```python
# X_Train
```

Learning Phase student 


```python
# adding Simple imputer to fever col
si = SimpleImputer()
X_Train_fever = si.fit_transform(X_Train[["fever"]])
X_Test_fever = si.fit_transform(X_Test[["fever"]])
```

```python
# X_Test_fever
```

```python
# Ordinal encoding 
oe = OrdinalEncoder(categories=[["Mild" ,"Strong"]])
X_Train_cough = oe.fit_transform(X_Train[["cough"]])
X_Test_cough = oe.fit_transform(X_Test[["cough"]])
```

```python
# X_Train_cough    
```

```python
# one hot encoding
ohe = OneHotEncoder()

X_Train_gender_city = ohe.fit_transform(X_Train[["gender" ,"city" ]]).toarray()
X_Test_gender_city = ohe.fit_transform(X_Test[["gender", "city"]]).toarray()
# X_Test_gender_city
```

```python
# Extracting Age column
X_Train_age = X_Train.drop(columns=["gender" , "city" , "fever" , "cough"]).values
X_Test_age = X_Test.drop(columns=["gender" , "city" , "fever" , "cough"]).values

X_Train_age.shape
```

```python
X_train_Transformed = np.concatenate((X_Train_age , X_Train_fever, X_Train_gender_city, X_Train_cough), axis=1)
X_test_Transformed = np.concatenate((X_Test_age , X_Test_fever, X_Test_gender_city, X_Test_cough), axis=1)

X_train_Transformed.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
  transformers=[
   ("tnf1" , SimpleImputer() , ["fever"] ) ,
   ("tnf2" ,OrdinalEncoder(categories=[["Mild" , "Strong"]]) , ["cough"] ) ,
   ("tnf3" , OneHotEncoder() , ["gender" , "city"])
  ] , remainder="passthrough"
 )
```
